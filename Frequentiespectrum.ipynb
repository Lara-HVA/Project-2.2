```python
#Importen van de benodigde packaged
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importen van de data
data = np.loadtxt('350Hz_2.csv', delimiter=';', dtype=np.float32)
print(data)
```

```python
#Omzetten van de data naar een numpy array
arr = data[:,1]
#Array omzetten zodat de golf om 0 heen gaat
arr = arr - (np.max(arr)+np.min(arr))/2
#Defineren van fs, t_tot en een lijst van t
fs = 23000
t_tot = 0.06
t = np.linspace(0,t_tot,1400)#fs*t_tot)
```

```python
#Plotten van de data
plt.figure(dpi=100)
plt.plot(t,arr, 'r-', ms=4.0, lw=1)
plt.grid()
plt.xlabel(r"Tijd $t$ [s]")
plt.ylabel(r"Amplitude [a.u.]")
```

```python
N = int(fs*t_tot) # Nsamples in signaal
T = t_tot/N # Sampling tijd

#Uit voeren van de fourier transformatie
fhat = np.fft.fft(arr)
x = np.linspace(0.0, 1.0/(2.0*T), N//2) # frequency axis
y = 2.0/N * np.abs(fhat[:N//2])
```

```python
#Plotten van de intensiteit van het frequentiespectrum 
plt.plot(x,y)
plt.xlabel('Frequentie $f$ [Hz]')
plt.ylabel("Intensiteit [a.u.]")
plt.xlim(0,1000)
```

```python
#Printen van de corresponderende frequentie bij de hoogste intensiteit
max_y = max(y)  # Find the maximum y value
max_x = round(x[y.argmax()],2)  # Find the x value corresponding to the maximum y value
print("Dit is de frequentie", max_x, "Hz.")
```
